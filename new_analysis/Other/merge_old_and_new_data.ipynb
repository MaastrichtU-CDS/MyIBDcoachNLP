```python
import pandas as pd

# Sample old and new dataframes (replace with your actual dataframes)
old_df = annotated_df_lg
new_df = new_sentence_df

# Function to assign Manual_Category from old_df to new_df
def assign_manual_category(new_df, old_df, max_index=5000):
    manual_categories = []  # List to store the assigned categories
    
    # Iterate through each new sentence (stop at max_index or end of new_df)
    for i, new_sentence in enumerate(new_df['Sentence']):
        if i >= max_index:  # Stop processing after the max_index
            break
        
        matching_category = None
        
        # Loop through old_df to find a match
        for _, old_row in old_df.iterrows():
            # Check if the beginning of the old sentence matches the new sentence's start
            if new_sentence.startswith(old_row['Sentence']):  # Only compare the first 10 characters
                matching_category = old_row['Manual_Category']
                
                # Print the matched sentences (for debugging)
                print(f"iteration: {i}")
                #print(f"The following two sentences are matched: \n{new_sentence}\n{old_row['Sentence']}")
                break  # Stop once a match is found
        
        # Append the matched category (or None if no match found)
        manual_categories.append(matching_category)
    
    # Fill remaining manual_categories with None if we didn't reach max_index
    while len(manual_categories) < len(new_df):
        manual_categories.append(None)
    
    # Add the Manual_Category to the new_df
    new_df['Manual_Category'] = manual_categories
    return new_df

# Assign the Manual_Category from old_df to new_df based on the starting sentence match
new_df = assign_manual_category(new_df, old_df, max_index=5000)

```
