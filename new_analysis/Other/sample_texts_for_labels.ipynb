# Randomly sample 100 texts (not sentences) and label the sentences split from them

```python
import pandas as pd
```

```python
sentence_df = pd.read_excel("..\\..\\dataset\\analysis_datasets\\translated_sentence_data.xlsx")
sentence_df
```

```python
import random
unique_message_ID = set(sentence_df["Message_ID"].to_list())
sample_message_ID = random.sample(list(unique_message_ID), 1000)
sample_messages_df = sentence_df[sentence_df["Message_ID"].isin(sample_message_ID)]
sample_messages_df.to_excel("..\\..\\dataset\\analysis_datasets\\sample_sentences_for_labeling_1000.xlsx")
```

```python
# get the list of the 100 and 1000 samples
df_100 = pd.read_excel("..\\..\\dataset\\analysis_datasets\\sample_sentences_for_labeling_100.xlsx", index_col=0)
sample_100_ids = df_100["Sentence_ID"].to_list()
df_1000 = pd.read_excel("..\\..\\dataset\\analysis_datasets\\sample_sentences_for_labeling_1000.xlsx", index_col=0)
sample_1000_ids = df_1000["Sentence_ID"].to_list()
```
