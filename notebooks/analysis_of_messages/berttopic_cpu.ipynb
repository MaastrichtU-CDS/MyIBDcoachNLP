```python
messages = df["Message"].tolist()
messages = messages[2:]
messages[:10]
```

```python

# filtering only the 'to client' messages
filtered_df_patient = df[df['richting (van of naar cliënt)'] == 'From client']
filtered_df_doc = df[df['richting (van of naar cliënt)'] == 'To client']
```

```python
filtered_messages_patient = filtered_df_patient["Message"].tolist()
filtered_messages_patient[:10]
```

```python
filtered_messages_doc = filtered_df_doc["Message"].tolist()
filtered_messages_doc[:10]
```

```python
# remove html tags from the messages

from bs4 import BeautifulSoup

def clean_html_with_soup(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

messages_pat =  [clean_html_with_soup(message) for message in filtered_messages_patient]
messages_doc =  [clean_html_with_soup(message) for message in filtered_messages_doc]


# check the messages containing html tages (now they should be not exist)
for message in messages_pat:
    if '<br' in message:
        print(message)

for message in messages_doc:
    if '<br' in message:
        print(message)
```

```python
# remove duplicate messages
print(f'Number of cleaned patient messages: {len(messages_pat)}')
unique_messages_pat = list(set(messages_pat))
print(f'Number of unique cleaned patient messages: {len(unique_messages_pat)}')


print(f'Number of cleaned doctor messages: {len(messages_doc)}')
unique_messages_doc = list(set(messages_doc))
print(f'Number of unique cleaned doctor messages: {len(unique_messages_doc)}')
```
