```python
import pandas as pd
# Import two df
doc_info = pd.read_csv("/workspace/mijnidbcoachnlp/data/result_data/doc_info_st.csv")
doc_info_merged = pd.read_csv("/workspace/mijnidbcoachnlp/data/result_data/doc_info_st_merged.csv")
```

```python
doc_info_merged
```

```python
merged_df = pd.concat(
    [doc_info.set_index("Document")[["Topic"]],
     doc_info_merged.set_index("Document")[["Topic"]]],
    axis=1
).reset_index()

merged_df.columns = ["Document", "Topic_doc_info", "Topic_doc_info_merged"]
```

```python
merged_df
```

```python
# Group by 'Topic_doc_info_merged' and list the corresponding 'Topic_doc_info' values
subgroups = (
    merged_df.groupby("Topic_doc_info_merged")["Topic_doc_info"]
    .apply(set)
    .reset_index()
)

# Rename columns for clarity
subgroups.columns = ["Topic_doc_info_merged", "Topic_doc_info_group"]
```
