## Method 1: SpaCy embeddings

```python
!pip install -U pip setuptools wheel

#CPU OPTION
# !pip install -U spacy
# GPU option
!pip install tf-keras
!pip install -U 'spacy[cuda11x]' 
!python -m spacy download nl_core_news_sm
```

```python
spacy_model = "nl_core_news_sm"
```

```python
import spacy
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']
nlp = spacy.load(spacy_model, exclude=['tagger', 'parser', 'ner', 'attribute_ruler', 'lemmatizer'])

topic_model = BERTopic(embedding_model=nlp)
topics, probs = topic_model.fit_transform(docs)

fig = topic_model.visualize_topics()
fig.show()
```

## Method 2:  Sentence transformers for embeddings

```python
embedding_model = "sentence-transformers/distiluse-base-multilingual-cased-v1" # multilingual model

from bertopic import BERTopic
topic_model = BERTopic(embedding_model=embedding_model)
```

```python
topics, probs = topic_model.fit_transform(messages)

fig = topic_model.visualize_topics()
fig.show()
```

```python
topics, probs = topic_model.fit_transform(messages_fromclient)

fig = topic_model.visualize_topics()
fig.show()
```

## Method 3 experimenting with other parameters

```python
from hdbscan import HDBSCAN
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model=CountVectorizer(ngram_range=(1, 3), stop_words="english")

cluster_model = KMeans(n_clusters=50)
hdbscan_model = HDBSCAN(min_cluster_size=15, metric='euclidean', cluster_selection_method='eom', prediction_data=True)

topic_model = BERTopic(embedding_model=embedding_model, vectorizer_model=vectorizer_model)
```

```python
topics, probs = topic_model.fit_transform(messages_fromclient)

fig = topic_model.visualize_topics()
fig.show()
```
