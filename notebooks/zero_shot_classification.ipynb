```python
import pandas as pd
# read the sentence data 
df = pd.read_excel("/workspace/persistent/mijnidbcoachnlp/data/analysis_data/sentence_data_for_analysis.xlsx", index_col=0)
sentences = df["sentence"].to_list()

# check the df head
df.head()
```

```python
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired

# We define a number of topics that we know are in the documents
zeroshot_topic_list = ["pijn", "test", "bloed", "vraag", "apotheek", "medicatie", "onderzoek", "afspraak", "vaccinatie", "huisarts", "zwanger", "formulier", "eten", "werk"]

# We fit our model using the zero-shot topics
# and we define a minimum similarity. For each document,
# if the similarity does not exceed that value, it will be used
# for clustering instead.
topic_model = BERTopic(
    embedding_model="sentence-transformers/distiluse-base-multilingual-cased-v1",
    min_topic_size=15,
    zeroshot_topic_list=zeroshot_topic_list,
    zeroshot_min_similarity=.85,
    representation_model=KeyBERTInspired()
)
topics, _ = topic_model.fit_transform(sentences)
```
