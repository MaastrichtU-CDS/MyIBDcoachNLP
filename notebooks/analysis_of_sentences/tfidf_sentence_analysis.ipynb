# BERTopic with TF-IDF imbeddings

```python
import pandas as pd
sentence_df = pd.read_excel("..\\..\\dataset\\analysis_datasets\\sentence_data.xlsx")
clean_sentences = sentence_df["Sentence_No_Greetings"][sentence_df["Sentence_No_Greetings"].str.len() > 20].to_list()
```

```python
### Importing the list of Dutch stopwords

with open('..\\..\\dataset\\stopwords.txt', 'r') as file:
    lines = [line.strip() for line in file.readlines()]

dutch_stopwords = lines
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN

# Create TF-IDF sparse matrix
docs = clean_sentences
vectorizer = TfidfVectorizer(min_df=5, stop_words=dutch_stopwords)
embeddings = vectorizer.fit_transform(docs)

# Train our topic model using TF-IDF vectors
topic_model_tf = BERTopic(

    # Pipeline models
    embedding_model=None,
    # Hyperparameters
    top_n_words=10,
    verbose=True
)
topics, probs = topic_model_tf.fit_transform(docs, embeddings)
```
